```python
import pandas as pd
x = [70,80,85]

df = pd.DataFrame(x)
df.describe()
```

```python
7.637626**2
```

```python
df.plot()
```

```python
df.plot.hist()
```

```python
df.plot.density()
```

```python
df.mode()
```

```python
ax = df.plot.density(title="Data Handling",figsize=(16,12),grid=True,legend=False)
ax.axvline(x=df.describe()[0][1],)
data = ax.get_lines()[0].get_data()
hd = data[1].max()
ax.text(df.describe()[0][5],hd*0.75,"mean")
ax.text(df.median()[0],hd*0.5,"median")
ax.text(df.mode().max()[0],hd*0.25,"mode")
ax.axhline(hd,color="red")
ax.text(-1,hd-0.01,f"Highest Density ({hd:0.3})")
print("Highest Density value",hd)
```
